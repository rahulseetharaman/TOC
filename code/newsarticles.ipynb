```python
import pandas as pd
import os
import re
```

```python
topics={'business':0,'entertainment':1,'politics':2,'sport':3,'tech':4}
```

```python
dataset=[]
labels=[]
```

```python
    def clean_data(string):
        string = re.sub(r"\'s", "", string)
        string = re.sub(r"\'ve", "", string)
        string = re.sub(r"n\'t", "", string)
        string = re.sub(r"\'re", "", string)
        string = re.sub(r"\'d", "", string)
        string = re.sub(r"\'ll", "", string)
        string = re.sub(r",", "", string)
        string = re.sub(r"!", " ! ", string)
        string = re.sub(r"\(", "", string)
        string = re.sub(r"\)", "", string)
        string = re.sub(r"\?", "", string)
        string = re.sub(r"'", "", string)
        string = re.sub(r"[^A-Za-z0-9(),!?\'\`]", " ", string)
        string = re.sub(r"[0-9]\w+|[0-9]","", string)
        string = re.sub(r"\s{2,}", " ", string)
        #r=Rake()
        #r.extract_keywords_from_text(string)
        #keywords=r.get_ranked_phrases()
        #print(keywords)
        #return ' '.join(keywords).strip().lower()
        return string.strip().lower()
```

```python
print(os.getcwd())
for topic in topics.keys():
    dirname='..\\dataset\\'+topic
    files=os.listdir(dirname)
    for file in files:
        with open(os.path.join(dirname,file),'r') as f:
            dataset.append(clean_data(f.read()))
            labels.append(topics[topic])
            
```

```python
dataset
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
vectorizer=TfidfVectorizer(stop_words='english',min_df=2)
```

```python
X_train,X_test,y_train,y_test=train_test_split(dataset,labels,test_size=0.2)
```

```python
vectors=vectorizer.fit_transform(X_train)
```

```python
clf=MultinomialNB()
clf.fit(vectors,y_train)
```
